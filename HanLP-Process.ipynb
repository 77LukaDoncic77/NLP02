```python
import hanlp
HanLP = hanlp.load(hanlp.pretrained.mtl.CLOSE_TOK_POS_NER_SRL_DEP_SDP_CON_ELECTRA_SMALL_ZH) # 世界最大中文语料库
# 读入文件
path = "C:/Users/24372/Desktop/学习/大三上/自然语言处理/NLP大作业/预处理后(干净).txt"
file = open(path,encoding="utf-8") # 打开待分词文本文件
file_data = file.readlines()

print("Results of Hanlp.tok/fine-精细版:\n")
for sentence in file_data:
    dic = HanLP(sentence)
    res = dic.setdefault("tok/fine")
    res = '/'.join(res)
    print(res)

print()
print("----------------------------分界线------------------------------\n")

print("Results of Hanlp.tok/coarse-粗糙版:\n")
for sentence in file_data:
    dic = HanLP(sentence)
    res = dic.setdefault("tok/coarse")
    res = '/'.join(res)
    print(res)
```

```python
# 统计测试集总分词数
res_path = "hanlp_fine.txt"
res_file = open(res_path,encoding="utf-8") # 打开分词结果文本文件
res_file_cen = res_file.readlines()
res_len = 0
for cen in res_file_cen:
    res_len += len(cen)
print(res_len)
```

```python
# 统计标准答案总分词数
source_path = "预处理后(正确分词).txt"
source_file = open(source_path,encoding="utf-8") # 打开分词结果文本文件
source_file_cen = source_file.readlines()
source_len = 0
for cen in source_file_cen:
    source_len += len(cen)
print(source_len)
```

```python
# 统计测试结果中正确的分词数
right = 0
for res_cen, source_cen in zip(res_file_cen, source_file_cen):
    for res_word in res_cen:
        if res_word in source_cen:
            right += 1
print(right)
```

```python
# 打印结果
p = (right / res_len) * 100
r = (right / source_len) * 100
f1 = (2 * p * r) / (p + r)
print("precession:",p)
print("Recall:",r)
print("F1-score:",f1)
```

```python
import csv

# 将词保存成集合形式
def getWord(str):
    list = []
    s = 0
    for word in str.split("/"):
        e = s + len(word)
        list.append((s, e - 1))
        s = e
    return set(list)

# ori是原来的分词，pred是系统预测的输出
def evaluate(ori,pred):
    predSize = len(getWord(pred))
    oriSize = len(getWord(ori))
    rightSize = len(getWord(ori) & getWord(pred))
    print("predsize:",predSize)
    print("oriSize:",oriSize)
    print("rightSize:",rightSize)
    # Recall
    R = rightSize / oriSize
    # Precision
    P = rightSize / predSize
    # F-measure
    if P == 0 or R == 0:
        return R, P, 'invalid'
    F = 2 * P * R / (P + R)
    return R, P, F

if __name__ == '__main__':
    # pred_1 = "终场/前/2/分钟/解放/军队/射门/被/朱力/挡出/,/一/记/长传/,/9/号/穆瑞正/个人/突破/再/添/一/分/。"
    # pred_2 = "终场/前/２分钟/解放/军队/射门/被/朱力/挡/出/,/一/记/长传/,/９号/穆瑞正/个人/突破/再/添/一/分/。"
    # ori = "终场/前/2/分钟/解放/军队/射门/被/朱力/挡/出/,/一/记/长传/,/9/号/穆瑞正/个人/突破/再/添/一/分/。"
    #with open('Results-of-HanLP(粗糙版).csv', 'w', newline="") as f0:
    with open('Results-of-HanLP(精细版).csv', 'w', newline="") as f0:
        csv_write = csv.writer(f0)
        csv_write.writerow(["Recall", "Precision", "F-measure"])
        with open('预处理后(正确分词).txt', 'r', encoding='utf-8') as f:
            #with open('hanlp_coarse.txt', 'r', encoding='utf-8') as f1:
            with open('hanlp_fine.txt', 'r', encoding='utf-8') as f1:
                a = f.readlines()
                b = f1.readlines()
                total_R = 0.0
                total_P = 0.0
                total_F = 0.0
                for i in range(0, len(a)):
                    pred = b[i]
                    ori = a[i]
                    print(i + 1, '\t', evaluate(ori,pred))
                    R, P, F = evaluate(ori,pred)
                    total_R += R
                    total_P += P
                    total_F += F
                    csv_write.writerow([R, P, F])
                csv_write.writerow("")
                csv_write.writerow(["total_recall", "total_precision", "total_F-measure"])
                total_R /= len(a)
                total_P /= len(a)
                total_F /= len(a)
                csv_write.writerow([total_R, total_P, total_F])
                print(len(a) + 1, '\t', (total_P, total_R, total_F))
    # print(evaluate(ori, pred_1))
    # print(evaluate(ori,pred_2))
```
